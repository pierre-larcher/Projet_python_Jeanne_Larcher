```python
def ordre(premier,dernier):
    sens = 0
    if int(premier) < int(dernier):
        sens = 'normal'
    else: 
        sens = 'inverse'
    return sens
    
sens=ordre(premier,dernier)
```

```python
# Calcul du chemin minimal

def calcule_temps(graphe: a1, depart: str, arrivee: str) -> float:
    d0 = {sommet: float("inf") for sommet in graphe.sommets()}
    d0[depart] = 0.0
    d1 = _calcule_suivante(graphe, depart, d0)
    while d1 != d0:
        d0, d1 = d1, _calcule_suivante(graphe, depart, d1)
    return d1,d1[arrivee]
```

```python
# On réinject pour avoir les candidat à la minimisation

def _recuperation_candidat_min_chemin(graphe: a1, depart: str, temps: dict[str, float]) -> dict[str, float]:
    candidat = list()
    for sommet in graphe.sommets():
        for predecesseur,poids in graphe[sommet]:
            candidat.extend([sommet,predecesseur, temps[predecesseur], poids])
    return candidat

result2=_recuperation_candidat_min_chemin(a1,premier,result[0])
result2
```

```python
def trouve_chemin_minimal(candidat):
    
    newlist = list()
    
    for i in candidat:
        newlist.append(sum(i[-2:]))
    id=(newlist).index(min(newlist))
    
    mini = candidat[id]
    
    return mini

type(trouve_chemin_minimal(minim[0]))
```

```python
def aa(liste):
    liste_fi = list()
    for i in liste:
        a=trouve_chemin_minimal(i)
        liste_fi.append(a)
    return liste_fi

ok=aa(minim)
ok
```
